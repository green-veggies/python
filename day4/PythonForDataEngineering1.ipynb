## Assignment - Split the dataset based on the stock symbol and write the records of each stock into an individual output file named <stockname>.csv.

```python
with open('nyse_sample_data.csv','r') as f:
  line_str = f.read()
  lines = line_str.splitlines()

  print(lines)

  set1 = set()
  set1 = set(map(lambda x:x.split(',')[1],lines))


  set1.remove('stock')

  for x in set1:
    with open(x+'.csv','a') as f:
      for line in lines:
        if x==line.split(',')[1]:
          f.write(line + '\n')
```

```python
import pandas as pd

dataPath = 'sample_data.csv'
# print(dataPath)

df = pd.read_csv(dataPath)
# print(df.tail())
# print(df.columns)

orders_columns = ['order_id','order_date','order_customer_id','order_status']
df = pd.read_csv(dataPath,names = orders_columns)
# print(df.head())
df['order_status'].unique()

df.query('order_status == ("COMPLETE","CLOSED")')
```

## Count of each order status

```python
df.groupby('order_status')['order_id'].agg(order_count='count')
```
