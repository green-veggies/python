## Pandas exercise

```python
import pandas

data_path = '/content/sales_2024-09-01.csv'
df = pandas.read_csv(data_path)

sales_dic = {}
for store_id,sales_revenue in zip(df['store_id'],df['dollar_sales']):
    if store_id in sales_dic:
        sales_dic[store_id] += sales_revenue
    else:
        sales_dic[store_id] = sales_revenue
sorted_data = sorted(sales_dic, key = sales_dic.get)[-1:-4:-1]

# print(sorted_data[-1:-4:-1])

for i in sorted_data:
  print(i,sales_dic[i])



```

```python
store_rev = (
    df.groupby(["store_id", "store_name"], as_index=False)["dollar_sales"]
      .sum()
      .rename(columns={"dollar_sales": "total_revenue"})
)

top3 = (
    store_rev.sort_values("total_revenue", ascending=False)
             .head(3)
             .reset_index(drop=True)
)

top3
```

## API Integration of open weather api

```python
import requests
from datetime import datetime

LAT, LON = 12.9716, 77.5946  # Coordinates for Bengaluru
LAT, LON


unix_timestamp = int(datetime.now().timestamp())
unix_timestamp


API_KEY = "5883b9ac08253fbd080c073b1832a6a8"


url = (
    "https://api.openweathermap.org/data/3.0/onecall/timemachine"
    f"?lat={LAT}&lon={LON}&dt={unix_timestamp}&appid={API_KEY}&units=imperial"
)
url


response = requests.get(url)
response.status_code

weather_data = []

if response.status_code == 200:
    data = response.json()
    today = datetime.now().date()

    day_weather = {
        "date": today,
        "temp": data["data"][0]["temp"],
        "humidity": data["data"][0]["humidity"],
        "weather_main": data["data"][0]["weather"][0]["main"],
        "weather_desc": data["data"][0]["weather"][0]["description"],
    }
    weather_data.append(day_weather)
else:
    print("Failed to fetch data for today")

weather_data
```
